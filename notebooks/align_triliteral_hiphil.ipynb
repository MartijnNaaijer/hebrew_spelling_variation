```python
import collections
import os

import pandas as pd
```

```python
dat = pd.read_csv(os.path.join('../data', 'hiphil_triliteral_with_hireq.csv'), sep='\t')
dat.shape
```

```python
mt = dat[dat.scroll == 'MT']
mt.shape
```

```python
dss = dat[dat.scroll != 'MT'] 
dss.shape
```

```python
mt.head()
```

```python
mt_hireq = mt[mt.has_hireq == '1']
mt_hireq.shape
```

```python
dss_hireq = dss[dss.has_hireq == '1']
dss_hireq
```

```python
dss_match_ids = [int(dss_id) for dss_id in dss_hireq.mt_match.to_list()]
dss_match_ids
```

```python
full_count = collections.defaultdict(int)

for _, row in mt_hireq.iterrows():
    tf_id = row.tf_id
    has_vowel = row.has_vowel_letter
    bo, ch, ve = row.book, row.chapter, row.verse
    lex = row.lex
    g_cons = row.g_cons

        
    #print(tf_id, row.book)
    if tf_id in dss_match_ids:
        dss_relevant = dss_hireq[dss_hireq.mt_match == str(tf_id)]
        #print(dss_relevant.shape[0])
        for _, dss_row in dss_relevant.iterrows():
            has_vowel_dss = dss_row.has_vowel_letter
            scroll = dss_row.scroll
            g_cons_dss = dss_row.g_cons
            
            full_count[(has_vowel_dss, has_vowel)] += 1
            if has_vowel != has_vowel_dss:
                print(dss_relevant.shape[0], tf_id, bo, ch, ve, lex, g_cons, g_cons_dss, has_vowel, has_vowel_dss, scroll)
            #if has_vowel == 0 and has_vowel_dss == 0:
            #    print(dss_relevant.shape[0], tf_id, bo, ch, ve, lex, g_cons, g_cons_dss, has_vowel, has_vowel_dss, scroll)
            
print(full_count)
```
