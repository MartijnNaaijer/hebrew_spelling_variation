```python
# words with multiple matres (but no double mater MM)

words_with_multip_matres = 0

sum([1 if pattern.replace('MM', 'M').count('M') > 1 else 0 for pattern in dat.pattern_g_cons if isinstance(pattern, str)])
```

```python
for tfid, pat, gc in zip(dat.tf_id, dat.pattern_g_cons, dat.g_cons):
    if isinstance(pat, str):
        if pat.replace('MM', 'M').count('M') > 2:
            print(tfid, gc, pat.replace('MM', 'M').count('M'))
```

```python
sum([1 if pattern.endswith('M') else 0 for pattern in dat.pattern_g_cons if isinstance(pattern, str)])
```

```python
# is final consonantal H correctly classified as C? Answer: yes.

for tfid, pat, gc in zip(dat.tf_id, dat.pattern_g_cons, dat.g_cons):
    if isinstance(pat, str):
        if gc[-1] == 'H' and pat[-1] == 'C' and len(pat) > 1:
            print(tfid, gc, pat)
```

```python
# medial matres

sum([pattern.replace('MM', 'M').rstrip('M').count('M') for pattern in dat.pattern_g_cons if isinstance(pattern, str)])
```

```python
# How many verses and verses without matres.

verses_matres = collections.defaultdict(int)

for bo, ch, ve, pat in zip(dat.book, dat.chapter, dat.verse, dat.pattern_g_cons):
    if isinstance(pat, str):
        verses_matres[(bo, ch, ve)] += pat.count('M')
```
